3. Astronomer Rees argues that this data could be fit by the function $y =  Asin(Bx)+C$. Plot this function on top of your data with some guesses at what $A$, $B$ and $C$ are (hint: $C=0.5$ provides a good fit to the data).<br><br>

### Answer:

*Your answer here:*

```python
from numpy import array,sin
from matplotlib.pyplot import xlabel,ylabel,title,legend

#question data
time = [0.,0.08163265,0.24489796,0.32653061,0.48979592,
        0.73469388,0.81632653,0.97959184,1.2244898,1.30612245,
        1.46938776,1.71428571,1.79591837,1.95918367,2.20408163,
        2.28571429,2.44897959,2.7755102,2.93877551,3.18367347,
        3.26530612,3.42857143,3.67346939,3.75510204,4.]

#converting from list to array
x_data=array(time)

#question data
brightness = [0.36253211,1.16748232,1.78775018,2.01934344,2.70657393
              ,2.57897185,2.96562829,3.00174824,2.93663326,
              2.80591525,2.11027035,1.67940753,1.64772549,0.3513817,0.33366962,0.47097419,-0.39790443,-0.65133529,-1.52086508,-1.76831912
              ,-2.05865798,-1.55083722,-1.90755375,-1.74307249,-1.66250842]

#converting from list to array
y_data=array(brightness)

#plotting data with errorbars, error in brightness is given to be 10% of the value
plt.errorbar(x_data,y_data,yerr=abs(0.1*y_data),elinewidth=5,fmt="gD",label="Astronomical data")
xlabel("Time (days)")
ylabel("Brightness (fraction of normal brightness)")
title("Graph of brightness as a fraction of normal brightness\n against time in days for supernova explosion")

#creating fitting function values
A=2.6
B=1.4
C=0.5
yfit=A*sin(B*x_data)+C
#plotting the proposed fitting function
plt.plot(x_data,yfit,label="fitting function")
legend()
```

The fitting function fits the data well around 1 day and sits inside the acceptable errors. The function is not inside the acceptable errors from 0 to 0.5 days which implies the supernova was increasing in brightness faster than the model could account for. Around 1.5 days the function does not fit the data well, implying that the supernova was decreasing in brightness faster than the model could account for. due to the brightness being close to 0 around 2 days, it is hard to say if the model is a good fit as the errors are very small at this time. Between 3 and 4 days the model seems to fit the data well, with notable exceptions at 3.5 days and 4 days. These may be anomalies or it may imply that the model is not well suited as the time progresses.
